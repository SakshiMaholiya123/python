```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,5,11)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("graph")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

subpolts in matplotlib - creating multiple plots 

```python
plt.subplot(2,2,1)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

object oriented way- to make changes in the canvas(white background)

dpi(dots per inch)=to change the pixcel of the graph  and figsize= to change the figure size

```python
fig=plt.figure(figsize=(6,8),dpi=500)
axis_1=fig.add_axes([0.1,0.1,0.8,0.8])
axis_1.plot(x,y)
axis_2=fig.add_axes([0.2,0.2,0.4,0.4])
axis_2.plot(x,y)
```

types of plot

scatter plot = used  to create regression algorithm

```python
plt.scatter(x,y)
```

histogram

```python
plt.hist(y)
```

boxplot

```python
plt.boxplot(y)
```

saving plots

```python
fig=plt.figure()
ax=fig.add_axes([0.1,.1,.8,.8])
ax.plot(x,y)
plt.savefig("plot.jpg")
```
