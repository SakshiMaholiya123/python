```python
file.replace(to_replace=21,value=20)
```

```python
file.replace([1,2,3,4,5],25)
```

Replace using regex

```python
file.replace("[A-Za-z]",'python',regex=True)
```

Replace using dictionary

```python
file.replace({'Name':'[A-Z]'},22,regex=True)
```

```python
import numpy as np
file.replace(52,np.nan).ffill()
```

```python
file.replace(1023,np.nan).bfill()
```

Repace using limit - not present in newer version of pandas(gives error)

```python
file.replace(22,'sakshi',limit=4)
```

```python
file.replace(22,'sakshi',inplace=True)
```

interpolate[mainly works with numeric column]-it is used to fill the missing data ,not randomly but after the obseving its sequence values

```python
csv_file=pd.read_csv('students_missing_data.csv')
```

```python
csv_file
```

```python
csv_file["Marks"]=csv_file["Marks"].interpolate()

csv_file["Age"]=csv_file["Age"].interpolate()
```

```python
csv_file
```

```python
csv_file=csv_file['Age'].interpolate(method='quadratic')
```

```python
csv_file
```

```python
csv_file["Age"] = csv_file["Age"].interpolate(method='linear',limit=1)
```

```python
csv_file
```

```python
csv_file['Marks'].interpolate(limit_direction='forward',limit=2)
```

```python
csv_file['Age']=csv_file['Age'].interpolate(method='linear',limit=1,limit_direction='both')
```

```python
print(csv_file)
```
