```python
import pandas as pd
```

```python
data1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
data2=pd.DataFrame({'A':[1,2,3,5],'c':[15,16,17,18]})
```

```python
pd.merge(data1,data2,on='A')
```

```python
pd.merge(data1,data2,how='left')
```

```python
pd.merge(data1,data2,how='inner')
```

```python
pd.merge(data1,data2,how='outer',indicator=True)
```

```python
pd.merge(data1,data2,left_index=True,right_index=True,suffixes=('id_1','id'))
```

concat

```python
s1=pd.Series([1,2,3,4])
s2=pd.Series([11,12,13,14])
```

```python
s1
```

```python
s2
```

```python
pd.concat([s1,s2])
```

```python
d1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
d2=pd.DataFrame({'A':[1,2,3,5],'c':[15,16,17,18]})
```

```python
pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],axis=1)
```

concat using join parameter


```python
pd.concat([d1,d2],join='outer')
```

```python
pd.concat([d1,d2],join='inner')
```

concat using key parameter


```python
pd.concat([d1,d2],keys=['d1','d2'])
```

```python
pd.concat([d1,d2],keys=['d1','d2'],axis=1)
```
