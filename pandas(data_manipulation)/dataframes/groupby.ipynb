groupby-it helps us to make the data in sequence in the datasheets

```python
import pandas as pd
```

```python
data=pd.DataFrame({'name':['a','b','c','d','e','a','b','g','c'],
                   's1_marks':[81,72,53,94,95,81,72,86,53],
                   's2_marks':[87,65,67,98,67,87,65,67,67]})
```

```python
data
```

```python
data_new=data.groupby('name')
data_new
```

```python
for i,j in data_new:
    print(i)
    print(j)
```

```python
data_new.get_group(name='a')
```

```python
data_new.min()
```

```python
data_new.max()
```

```python
data_new.mean()
```

```python
data_new.sum()
```

```python
data_new.agg(['min','max','sum','median'])
```
