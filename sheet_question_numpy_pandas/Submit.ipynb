```python
file['Sustainability_Rank']=pd.to_numeric(file['Sustainability_Rank'],errors='coerce')
file['Faculty_Student_Rank']=pd.to_numeric(file['Faculty_Student_Rank'],errors='coerce')
file['International_Faculty_Rank']=pd.to_numeric(file['International_Faculty_Rank'],errors='coerce')
file['Academic_Reputation_Rank']=pd.to_numeric(file['Academic_Reputation_Rank'],errors='coerce')
file['Employer_Reputation_Rank']=pd.to_numeric(file['Employer_Reputation_Rank'],errors='coerce')
file['Citations_per_Faculty_Rank']=pd.to_numeric(file['Citations_per_Faculty_Rank'],errors='coerce')
file['International_Students_Rank']=pd.to_numeric(file['International_Students_Rank'],errors='coerce')
file['International_Research_Network_Rank']=pd.to_numeric(file['International_Research_Network_Rank'],errors='coerce')
file['Employment_Outcomes_Rank']=pd.to_numeric(file['Employment_Outcomes_Rank'],errors='coerce')
file['Overall_Score']=pd.to_numeric(file['Overall_Score'],errors='coerce')
```

```python
file_drp=file.dropna()
file_drp
```

```python
file['Score_Category']=np.where(file['Overall_Score']>=90,'Elite',
                            np.where(file['Overall_Score']>=75,'High',
                                np.where(file['Overall_Score']>=50,'Medium',
                                    'Low',)))

```
